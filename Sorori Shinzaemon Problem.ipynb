**[QUESTION 1]: Number of rice grains on the 100th day**

```python
import matplotlib.pyplot as plt
import numpy as np

total_squares = 64

grains_per_square = []
cumulative_grains = []

# calculating the total amount of grains gained over 100 days
total_grains = 0
for square in range(total_squares):
    grains = 2 ** square
    total_grains += grains
    grains_per_square.append(grains)
    cumulative_grains.append(total_grains)

total_grains_str = f"{total_grains:,}"
print(f"Total number of grains of rice: {total_grains_str}")

x = np.arange(1, total_squares + 1)

plt.figure(figsize=(12, 6))

# plot grains per square
plt.subplot(1, 2, 1)
plt.plot(x, grains_per_square, label='Grains per Square', color='blue')
plt.yscale('log')
plt.title('Grains of Rice per Square')
plt.xlabel('Square Number')
plt.ylabel('Grains of Rice (Log Scale)')
plt.grid(True)
plt.legend()

# plot cumulative grains
plt.subplot(1, 2, 2)
plt.plot(x, cumulative_grains, label='Cumulative Grains', color='green')
plt.yscale('log')
plt.title('Cumulative Grains of Rice')
plt.xlabel('Square Number')
plt.ylabel('Cumulative Grains of Rice (Log Scale)')
plt.grid(True)
plt.legend()

# show the plots
plt.tight_layout()
plt.show()
```

**[QUESTION 2]: Number of rice grains other than the 100th day**

```python
import matplotlib.pyplot as plt
def compute_sorori_shinzaemon(n_days = 100):

  list_n_grains = []
  list_total_grains = []
  total_grains = 0

  for day in range(n_days):
      grains = 2 ** day
      total_grains += grains
      list_n_grains.append(grains)
      list_total_grains.append(total_grains)

  return list_n_grains, list_total_grains

# number of days for the course
course_days = 10
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days = course_days)

# display the total grains of rice for the course
total_grains_course = list_total_grains[-1]
print(f"Total grains of rice received over {course_days} days: {total_grains_course:,}")

# plotting the data
plt.subplot(1, 2, 1)
plt.plot(range(1, course_days + 1), list_n_grains, label ='Grains per day', color ='blue')
plt.yscale('log')
plt.title('Grains of Rice per Day')
plt.xlabel('Day')
plt.ylabel('Grains of Rice (Log Scale)')
plt.grid(True)
plt.legend()

# plot cumulative grains
plt.subplot(1, 2, 2)
plt.plot(range(1, course_days + 1), list_total_grains, label ='Cumulative Grains', color ='green')
plt.yscale('log')
plt.title('Cumulative Grains of Rice')
plt.xlabel('Day')
plt.ylabel('Cumulative Grains of Rice (Log Scale)')
plt.grid(True)
plt.legend()

# show the plots
plt.tight_layout()
plt.show()
```

**[QUESTION 3]: How many people can survive on the rice they receive for how many days?**

```python
import matplotlib.pyplot as plt
def compute_sorori_shinzaemon(n_days=100):

    list_n_grains = []
    list_total_grains = []
    total_grains = 0

    for day in range(n_days):
        grains = 2**day
        total_grains += grains
        list_n_grains.append(grains)
        list_total_grains.append(total_grains)

    return list_n_grains, list_total_grains

def calculate_survival_days(total_grains, num_people):

    RICE_PER_PERSON_PER_DAY = 2000  # ASSUMPTION: 2000 grains of rice needed per person per day equivalent to 40g knowing that one grain weights 0.02g
    total_rice_needed_per_day = num_people * RICE_PER_PERSON_PER_DAY
    survival_days = total_grains // total_rice_needed_per_day
    return survival_days

# number of days for the course
course_days = 10
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=course_days)

# display the total grains of rice for the course
total_grains_course = list_total_grains[-1]
print(f"Total grains of rice received over {course_days} days: {total_grains_course:,}")

# calculate survival days
num_people_in_class = 20
survival_days = calculate_survival_days(total_grains_course, num_people_in_class)
print(f"{num_people_in_class} people can survive for {survival_days} days on the rice received over {course_days} days.")

# plotting the data
plt.figure(figsize=(12, 6))

# plot grains per day
plt.subplot(1, 2, 1)
plt.plot(range(1, course_days + 1), list_n_grains, label='Grains per Day', color='blue')
plt.yscale('log')
plt.title('Grains of Rice per Day')
plt.xlabel('Day')
plt.ylabel('Grains of Rice (Log Scale)')
plt.grid(True)
plt.legend()

# plot cumulative grains
plt.subplot(1, 2, 2)
plt.plot(range(1, course_days + 1), list_total_grains, label='Cumulative Grains', color='green')
plt.yscale('log')
plt.title('Cumulative Grains of Rice')
plt.xlabel('Day')
plt.ylabel('Cumulative Grains (Log Scale)')
plt.grid(True)
plt.legend()

# show the plots
plt.tight_layout()
plt.show()
```
