**[PROBLEM 1]: Implementation using arithmetic operators of power**

```python
thickness = 0.00008

folded_thickness = thickness * 2 ** 43

print("Thickness: {} meters".format(folded_thickness))
```

**[PROBLEM 2]: Unit conversion**

```python
print("Thickness: {} kilometers".format(folded_thickness/1000))
```

**[PROBLEM 3]: Create using a for statement**

```python
number_of_folds = 43
current_thickness = thickness
for i in range(number_of_folds):
    current_thickness *= 2
print("Thickness: {} meters".format(current_thickness))
print("Thickness: {} kilometers".format(current_thickness/1000))
```

**[PROBLEM 4]: Comparison of calculation times**

```python
import time
start = time.time()

number_of_folds = 43
current_thickness = thickness

for i in range(number_of_folds):
    current_thickness *= 2

elapsed_time = time.time() - start
print("Elapsed time: {}[s]".format(elapsed_time))
```

**[PROBLEM 5]: Saving to a list**

```python
thickness_list = []
thickness_list.append(thickness)

for i in range (number_of_folds):
    current_thickness *= 2
    thickness_list.append(current_thickness)

print(f"Number of values in the list: {len(thickness_list)}")
print(thickness_list)
```

**[PROBLEM 6]: Displaying a line graph**

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(thickness_list)
plt.show()
```

**[PROBLEM 7]: Customizing graphs**

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=20)
plt.plot(thickness_list, color='red')
plt.show()
```
