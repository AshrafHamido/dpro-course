**[Problem 1]: Creating random numbers**

```python
import numpy as np
import matplotlib.pyplot as plt

# Mean vector
mean = [-3, 0]

# Covariance matrix (example values)
covariance_matrix = [[1, 0.8], [0.8, 1]]

# Generate 500 random samples
data = np.random.multivariate_normal(mean, covariance_matrix, size=500)

# Plot the data
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], alpha=0.8)
plt.title("2D Normal Distribution with Mean (-3, 0)")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.axis("equal")
plt.show()
```

**[Problem 3]: Visualization using histograms**

```python
import numpy as np
import matplotlib.pyplot as plt

# Mean vector
mean = [-3, 0]

# Covariance matrix
covariance_matrix = [[1, 0.8], [0.8, 1]]

# Generate 500 random samples
data = np.random.multivariate_normal(mean, covariance_matrix, size=500)

# Determine the range for x-axis alignment
x_min = min(data[:, 0].min(), data[:, 1].min())
x_max = max(data[:, 0].max(), data[:, 1].max())

# Create histograms for X and Y dimensions
plt.figure(figsize=(12, 6))

# Histogram for the X-axis
plt.subplot(1, 2, 1)
plt.hist(data[:, 0], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histogram of X-axis")
plt.xlabel("X values")
plt.ylabel("Frequency")
plt.grid(True)
plt.xlim(x_min, x_max)

# Histogram for the Y-axis
plt.subplot(1, 2, 2)
plt.hist(data[:, 1], bins=30, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title("Histogram of Y-axis")
plt.xlabel("Y values")
plt.ylabel("Frequency")
plt.grid(True)
plt.xlim(x_min, x_max)

# Display the histograms
plt.tight_layout()
plt.show()
```

**[Problem 6]: Labeling**

```python
import numpy as np

# Data for Question 1
mean1 = [-3, 0]
covariance_matrix1 = [[1, 0.8], [0.8, 1]]
data1 = np.random.multivariate_normal(mean1, covariance_matrix1, size=500)

# Data for Question 4
mean2 = [0, -3]
covariance_matrix2 = [[1.0, 0.8], [0.8, 1.0]]
data2 = np.random.multivariate_normal(mean2, covariance_matrix2, size=500)

# Combine data
combined_data = np.vstack((data1, data2))

# Create labels
labels = np.concatenate((np.zeros(500), np.ones(500)))

# Add labels as a new column
labeled_data = np.column_stack((combined_data, labels))
labeled_data.shape
print(labeled_data)
```
