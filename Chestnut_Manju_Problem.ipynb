```python
import math
import matplotlib.pyplot as plt

# General function to calculate time to fill a space with growing objects
def calculate_time_to_fill(initial_volume, target_volume, doubling_time):
    current_volume = initial_volume
    time = 0

    while current_volume < target_volume:
        current_volume *= 2
        time += doubling_time

    return time

# Function to simulate and visualize the growth
def simulate_growth(initial_volume, target_volume, doubling_time):
    volumes = []
    times = []

    current_volume = initial_volume
    time = 0

    while current_volume < target_volume:
        volumes.append(current_volume)
        times.append(time)
        current_volume *= 2
        time += doubling_time

    # Append the final step when the target volume is exceeded
    volumes.append(current_volume)
    times.append(time)

    # Plot the growth
    plt.figure(figsize=(10, 6))
    plt.plot(times, volumes, label="Volume Growth", color="blue")
    plt.axhline(y=target_volume, color="red", linestyle="--", label="Target Volume")
    plt.xlabel("Time (minutes)")
    plt.ylabel("Volume (cubic meters)")
    plt.title("Exponential Growth of Volume Over Time")
    plt.legend()
    plt.yscale("log")
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.show()

# Example calculations
if __name__ == "__main__":
    # Example 1: Filling the solar system with chestnut buns
    CHESTNUT_BUN_VOLUME = 0.00025  # Approx. 250 cm^3 in cubic meters
    SOLAR_SYSTEM_VOLUME = 1.412e39  # Approx. volume of the solar system in cubic meters
    DOUBLING_TIME = 5  # Assume doubling every 5 minutes

    time_to_fill_solar_system = calculate_time_to_fill(CHESTNUT_BUN_VOLUME, SOLAR_SYSTEM_VOLUME, DOUBLING_TIME)
    print(f"Time to fill the solar system with chestnut buns: {time_to_fill_solar_system} minutes")
    simulate_growth(CHESTNUT_BUN_VOLUME, SOLAR_SYSTEM_VOLUME, DOUBLING_TIME)

    # Example 2: Filling the Tokyo Dome with balls
    BALL_VOLUME = 0.00419  # Approx. volume of a baseball in cubic meters
    TOKYO_DOME_VOLUME = 1.24e5  # Volume of Tokyo Dome in cubic meters

    time_to_fill_tokyo_dome = calculate_time_to_fill(BALL_VOLUME, TOKYO_DOME_VOLUME, DOUBLING_TIME)
    print(f"Time to fill the Tokyo Dome with balls: {time_to_fill_tokyo_dome} minutes")
    simulate_growth(BALL_VOLUME, TOKYO_DOME_VOLUME, DOUBLING_TIME)
```
