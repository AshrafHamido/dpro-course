**[Problem 3]: Finding the gradient**

```python
import numpy as np

# Create an ndarray for x from -50 to 50 with 0.1 intervals
x = np.arange(-50, 50.1, 0.1)

# Calculate y using the given equation
y = 0.5 * x + 1

# Calculate the gradient (dy/dx)
dy = np.diff(y)  # Change in y
dx = np.diff(x)  # Change in x

# Calculate the gradient as dy/dx
gradient = dy / dx

# Since gradient has one less element than x and y, we can create a new x array for gradients
x_gradient = x[:-1]  # Use x without the last element to match the gradient length

# Display the results
print("x coordinates for gradients:", x_gradient)
print("Gradients:", gradient)
```

**[Question 4]: Draw a graph**

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an ndarray for x from -50 to 50 with 0.1 intervals
x = np.arange(-50, 50.1, 0.1)

# Calculate y using the given equation
y = 0.5 * x + 1

# Calculate the gradient (dy/dx)
dy = np.diff(y)  # Change in y
dx = np.diff(x)  # Change in x
gradient = dy / dx

# Adjust x for gradients (removing the last element)
x_gradient = x[:-1]

# Plotting the linear function
plt.figure(figsize=(12, 6))

# First subplot for the linear function
plt.subplot(1, 2, 1)
plt.plot(x, y, '-r', label='y=0.5x + 1')
plt.title('Graph of y = 0.5x + 1')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend(loc='upper left')

# Second subplot for the gradient
plt.subplot(1, 2, 2)
plt.plot(x_gradient, gradient, '-b', label='Gradient')
plt.title('Graph of Gradient (dy/dx)')
plt.xlabel('x', color='#1C2833')
plt.ylabel('Gradient', color='#1C2833')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend(loc='upper left')

# Show the plots
plt.tight_layout()
plt.show()
```

**[Question 5]: Python function**

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_and_plot_gradients():
    # Define the ranges for x
    x1 = np.arange(-50, 50.1, 0.1)  # For y = x^2 and y = 2x^2 + 2^x
    x2 = np.arange(0, 50.1, 0.1)     # For y = sin(sqrt(x))

    # Define the equations
    y1 = x1**2
    y2 = 2 * x1**2 + 2**x1
    y3 = np.sin(np.sqrt(x2))

    # Calculate gradients
    dy1 = np.diff(y1)
    dx1 = np.diff(x1)
    gradient1 = dy1 / dx1

    dy2 = np.diff(y2)
    dx2 = np.diff(x1)
    gradient2 = dy2 / dx2

    dy3 = np.diff(y3)
    dx3 = np.diff(x2)
    gradient3 = dy3 / dx3

    # Ensure all gradients are of consistent shape for ndarray
    max_length = max(len(gradient1), len(gradient2), len(gradient3))

    # Pad gradients with NaN or zeros to make them equal length
    gradient1_padded = np.pad(gradient1, (0, max_length - len(gradient1)), 'constant', constant_values=np.nan)
    gradient2_padded = np.pad(gradient2, (0, max_length - len(gradient2)), 'constant', constant_values=np.nan)
    gradient3_padded = np.pad(gradient3, (0, max_length - len(gradient3)), 'constant', constant_values=np.nan)

    # Combine padded gradients into an ndarray
    gradients_ndarray = np.array([gradient1_padded, gradient2_padded, gradient3_padded])

    # Plotting
    plt.figure(figsize=(18, 6))

    # Plot for y = x^2
    plt.subplot(1, 3, 1)
    plt.plot(x1, y1, '-r', label='y=x^2')
    plt.title('Graph of y = x^2')
    plt.xlabel('x', color='#1C2833')
    plt.ylabel('y', color='#1C2833')
    plt.axhline(0, color='black', linewidth=0.5, ls='--')
    plt.axvline(0, color='black', linewidth=0.5, ls='--')
    plt.grid()
    plt.legend(loc='upper left')

    # Plot for y = 2x^2 + 2^x
    plt.subplot(1, 3, 2)
    plt.plot(x1, y2, '-g', label='y=2x^2 + 2^x')
    plt.title('Graph of y = 2x^2 + 2^x')
    plt.xlabel('x', color='#1C2833')
    plt.ylabel('y', color='yellow')
    plt.axhline(0, color='black', linewidth=0.5, ls='--')
    plt.axvline(0, color='black', linewidth=0.5, ls='--')
    plt.grid()
    plt.legend(loc='upper left')

    # Plot for y = sin(sqrt(x))
    plt.subplot(1, 3, 3)
    plt.plot(x2, y3, '-b', label='y=sin(sqrt(x))')
    plt.title('Graph of y = sin(sqrt(x))')
    plt.xlabel('x', color='orange')
    plt.ylabel('y', color='green')
    plt.axhline(0, color='black', linewidth=0.5, ls='--')
    plt.axvline(0, color='black', linewidth=0.5, ls='--')
    plt.grid()
    plt.legend(loc='upper left')

    # Show plots
    plt.tight_layout()
    plt.show()

    return gradients_ndarray

# Call the function and get gradients
gradients_result = compute_and_plot_gradients()

# Display the calculated gradients
print("Gradients ndarray:\n", gradients_result)
```

**[Problem 6]: Find the minimum value**

```python
import numpy as np
import matplotlib.pyplot as plt

def minimum_and_gradients():
    # Define the ranges for x
    x1 = np.arange(-50, 50.1, 0.1)  # For y = x^2 and y = 2x^2 + 2^x
    x2 = np.arange(0, 50.1, 0.1)     # For y = sin(sqrt(x))

    # Define the equations
    y1 = x1**2
    y2 = 2 * x1**2 + 2**x1
    y3 = np.sin(np.sqrt(x2))

    # Calculate gradients
    dy1 = np.diff(y1)
    dx1 = np.diff(x1)
    gradient1 = dy1 / dx1

    dy2 = np.diff(y2)
    dx2 = np.diff(x1)
    gradient2 = dy2 / dx2

    dy3 = np.diff(y3)
    dx3 = np.diff(x2)
    gradient3 = dy3 / dx3

    # Find minimum values and their indices
    min_y1 = np.min(y1)
    min_index1 = np.argmin(y1)

    min_y2 = np.min(y2)
    min_index2 = np.argmin(y2)

    min_y3 = np.min(y3)
    min_index3 = np.argmin(y3)

    # Output minimum values and indices
    print(f"Minimum of y=x^2: Value={min_y1}, at x={x1[min_index1]}")
    print(f"Minimum of y=2x^2 + 2^x: Value={min_y2}, at x={x1[min_index2]}")
    print(f"Minimum of y=sin(sqrt(x)): Value={min_y3}, at x={x2[min_index3]}")

    # Show gradients before and after the minimum for y=x^2
    gradient_before_min_1 = gradient1[min_index1 - 1] if min_index1 > 0 else "N/A"
    gradient_after_min_1 = gradient1[min_index1] if min_index1 < len(gradient1) else "N/A"

    print("Gradient before minimum (y=x^2):", gradient_before_min_1)
    print("Gradient after minimum (y=x^2):", gradient_after_min_1)

# Call the function to compute minimums and gradients
minimum_and_gradients()
```
